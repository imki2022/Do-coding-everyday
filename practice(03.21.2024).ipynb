# Natural Language Processing with PyTorch

```python
# pip install --upgrade scikit-learn

from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
```

```python
corpus = ['Time flies like an arrow.', 'Fruiot files lika a banana']

one_hot_vectorizer = CountVectorizer(binary = True) ## CountVectorizer는 텍스트 데이터를 단어의 빈도 표현으로 변환해주는 도구입니다. binary=True 파라미터를 사용하여 이진화된 단어 빈도를 생성하도록 설정합니다.

one_hot = one_hot_vectorizer.fit_transform(corpus).toarray() # fit_transform() 메서드를 사용하여 corpus를 피처 벡터로 변환합니다. toarray() 메서드를 사용하여 희소 행렬을 밀집 행렬로 변환합니다.

vocab = one_hot_vectorizer.get_feature_names_out() ## fit_transform() 메서드를 사용하여 corpus를 피처 벡터로 변환합니다. toarray() 메서드를 사용하여 희소 행렬을 밀집 행렬로 변환합니다.


sns.heatmap(one_hot, annot=True,
            cbar=False, xticklabels = vocab, yticklabels = ['Sentence 1', 'Sentence 2']) 

 ## seaborn 라이브러리를 사용하여 히트맵을 그립니다. annot=True로 설정하여 각 셀에 숫자를 표시하고, cbar=False로 설정하여 컬러바를 표시하지 않습니다.
# xticklabels와 yticklabels를 사용하여 x축과 y축에 각각 단어와 문장 레이블을 표시합니다.


# Countverctorizer 클래스는 기본적으로 글자 하나로 이루어진 단어를 무시합니다. 이 때문에 단어 'a'가 없습니다.
```

```python
# 한번 더

corpus = ['I want to graduate KAIST.', "Bayes Business School is not great School."]

one_hot_vectorizer = CountVectorizer(binary= True)

one_hot = one_hot_vectorizer.fit_transform(corpus).toarray()

vocab = one_hot_vectorizer.get_feature_names_out()


# sns.heatmap(one_hot, annot=True,            cbar=False, xticklabels = vocab, yticklabels = ['Sentence 1', 'Sentence 2']) 

sns.heatmap(one_hot, annot = True, cbar = False, xticklabels = vocab, yticklabels= ['sentence_1','sentence_2'])
```

##### TF_IDF - 흔한 토큰의 점수를 낮추고 드문 토큰의 점수를 높임

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns


tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(corpus).toarray()

sns.heatmap(tfidf, annot =True, cbar=False, xticklabels=vocab, yticklabels=['Sentence_1','Sentence_2'])
```
