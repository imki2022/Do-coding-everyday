```python
import pandas as pd

# Load the updated Excel file again to examine its contents
updated_data_path = 'data/Tikcer_09_03.xlsx'
updated_data = pd.read_excel(updated_data_path)
updated_data.head()
```

```python
# Group the data by 'Sector' and count the number of tickers in each sector
sector_counts = updated_data['Sector'].value_counts()
sector_counts
```

```python
import matplotlib.pyplot as plt
# Visualize the sector counts with annotations for the number of tickers
plt.figure(figsize=(10, 8))
bars = plt.bar(sector_counts.index, sector_counts.values, color='skyblue')

# Adding the count above the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center')

plt.title('Number of Tickers per Sector')
plt.xlabel('Sector')
plt.ylabel('Number of Tickers')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Calculate percentage and plot a pie chart
sector_percentage = sector_counts / sector_counts.sum() * 100

# Improve the readability of the pie chart by exploding the largest segments and adjusting the colors
explode_values = [0.1 if percentage > 10 else 0 for percentage in sector_percentage]  # Explode segments over 10%
colors = plt.cm.tab20c.colors  # More distinct color set

# Adjusting the layout to add space between the title and the chart
plt.figure(figsize=(10, 8))
plt.pie(sector_percentage, labels=sector_percentage.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode_values)
plt.title('Ticker Distribution by Sector as Percentage', pad=20)  # Increase padding for the title
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

```

```python
# Group the data by 'Sector' and count the number of tickers in each sector
exchange_counts = updated_data['Exchange'].value_counts()
exchange_counts
```

```python
# Calculate percentage and plot a pie chart
exchange_percentage = exchange_counts / exchange_counts.sum() * 100

plt.figure(figsize=(10, 8))
plt.pie(exchange_percentage, labels=exchange_percentage.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Ticker Distribution by Exchange as Percentage')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
