```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
etf = pd.read_csv("data/ETFs_daily.csv")

# Define functions
def SSD(x, y, init):
    return np.sum((np.log(x / init[0]) - np.log(y / init[1]))**2)

def SD(x, y, init):
    return np.std(np.log(x / init[0]) - np.log(y / init[1]))

def diff_pair(x, y, init):
    return np.log(x / init[0]) - np.log(y / init[1])

# Extract EWA and EWC data
EWA = etf[etf.iloc[:, 0] == etf.iloc[:, 0].unique()[0]]
EWC = etf[etf.iloc[:, 0] == etf.iloc[:, 0].unique()[1]]

# Calculate in-sample standard deviation
init_vals = [EWC.iloc[0, 3], EWA.iloc[0, 3]]
insample_sd = SD(EWC.iloc[:252, 3], EWA.iloc[:252, 3], init_vals)

# Calculate difference pair for in-sample and out-of-sample
in_diff_pair = diff_pair(EWC.iloc[:252, 3], EWA.iloc[:252, 3], init_vals)
out_diff_pair = diff_pair(EWC.iloc[252:378, 3], EWA.iloc[252:378, 3], init_vals)

# Replace NaN or Inf values with zeros
in_diff_pair = np.nan_to_num(in_diff_pair)
out_diff_pair = np.nan_to_num(out_diff_pair)

# Plot in-sample and out-of-sample difference pair
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(in_diff_pair)
plt.ylim(-np.nanmax(np.abs(in_diff_pair)), np.nanmax(np.abs(in_diff_pair)))
plt.axhline(y=2 * insample_sd, color='red', linestyle='--')
plt.axhline(y=-2 * insample_sd, color='red', linestyle='--')
plt.axhline(y=0, color='blue', linestyle='--')
plt.title('In-sample Difference Pair')

plt.subplot(2, 1, 2)
plt.plot(out_diff_pair)
plt.ylim(-np.nanmax(np.abs(out_diff_pair)), np.nanmax(np.abs(out_diff_pair)))
plt.axhline(y=2 * insample_sd, color='red', linestyle='--')
plt.axhline(y=-2 * insample_sd, color='red', linestyle='--')
plt.axhline(y=0, color='blue', linestyle='--')
plt.title('Out-of-sample Difference Pair')

plt.tight_layout()
plt.show()

# Calculate value difference price
value_diff_price = EWC.iloc[:, 3] - EWA.iloc[:, 3] * (EWC.iloc[0, 3] / EWA.iloc[0, 3])

# Plot value difference price
plt.figure(figsize=(8, 4))
plt.plot(value_diff_price)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Value Difference Price')
plt.xlabel('Time')
plt.ylabel('Value Difference Price')
plt.show()

# Initialize variables
N = len(EWC)
month = 21
year = 252
k = 5

X = EWC.iloc[:, 3]
Y = EWA.iloc[:, 3]

# Initialize arrays to store results
Return = np.zeros(int(N/month) - 18 + 1)
n_open = 0
n_sell = 0

for k in range(1, int(N/month) - 18 + 2):
    # Learning (12 month)
    L_period = range(month * (k - 1), month * (k - 1) + year)
    insample_sd = SD(X[L_period], Y[L_period], init=[X[L_period[0]], Y[L_period[0]]])
    
    # Trading period (6 month)
    T_period = range(max(L_period) + 1, max(L_period) + month * 6 + 1)
    out_diff_pair = diff_pair(X[T_period], Y[T_period], init=[X[L_period[0]], Y[L_period[0]]])
    
    index = np.where(np.abs(out_diff_pair) > 2 * insample_sd)[0]
    
    if len(index) != 0:
        signal_date = T_period[index[0]]
        n_open += 1
        
        pair_ind = out_diff_pair[index[0]] < 0
        ind1 = np.where(pair_ind, 1/X[signal_date + 1], 0) - np.where(~pair_ind, 1/Y[signal_date + 1], 0)
        ind2 = np.where(~pair_ind, 1/Y[signal_date + 1], 0) - np.where(pair_ind, 1/X[signal_date + 1], 0)
        
        W = np.array([ind1, ind2])
        port = np.dot(W, np.vstack((X[signal_date + 1: max(T_period)], Y[signal_date + 1: max(T_period)])))
        
        plt.plot(port)
        
        ind_sell = np.where(out_diff_pair[(index[0] + 1):] * out_diff_pair[(index[0] + 1)] <= 0)[0]
        
        if len(ind_sell) == 0:
            ret = port[-1]
        else:
            ret = port[min(ind_sell[0] + 1, len(port))]
            n_sell += 1
            print(k)
    else:
        ret = 0
        
    Return[k - 1] = ret

print("Number of sell signals:", n_sell)
print("Number of open positions:", n_open)
print("Returns:", Return)
print("Mean Return:", np.mean(Return))
```
